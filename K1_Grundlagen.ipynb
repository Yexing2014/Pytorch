## plot the relationship between w and loss

```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

```python
# function def
# forward function
def forward(x):
    return x * w 

def forward_3D(x):
    return x * W + B

# loss function
def loss_MSE(x,y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)

def loss_MSE_3D(x,y):
    y_pred = forward_3D(x)
    return (y_pred - y) * (y_pred - y)
```

### nur mit w

```python
# run, 2D
w_list = []
mes_list = []

w_arange = np.arange(0.0, 4.1, 0.1)

#each w
for w in w_arange:
    loss_sum = 0
    # sum of loss
    for x_val, y_val in zip(x_data, y_data):
        loss_val = loss_MSE(x_val, y_val)
        loss_sum += loss_val

    mes_val = loss_sum / 3
    w_list.append(w)
    mes_list.append(mes_val)
```

### mit w und b

```python
# run, 3D
w_arange = np.arange(0.0, 4.1, 0.1)
b_arange = np.arange(0.0, 4.1, 0.1)

W,B= np.meshgrid(w_arange,b_arange)

# sum of loss
for x_val, y_val in zip(x_data, y_data):
    loss_val = loss_MSE_3D(x_val, y_val)
    loss_sum += loss_val

mes_val = loss_sum / 3
```

### Plot the figure

```python
# plot 3D
fig = plt.figure()
ax = plt.axes(projection = '3d') 
ax.plot_surface(W,B,mes_val)
```
